## Renombrar variables

```python
# Para conocer el nombre de las columnas
datos01.columns
```

```python
# Manera 1 de renombrar: Especificar una nueva lista de nombres de columnas, previa copia
datos08 = datos01.copy()
datos08.columns = ['TOTAL_HOGAR', 'TOTAL_MUJERES', 'TOTAL_NINOS', 'ENTREVISTA',
                   'DEPARTAMENTO', 'AREA', 'NINOS_ELEG', 'HEMOG_ELEG', 'NOMCCPP',
                   'LONGITUDX', 'LATITUDY']
# debemos indicar la cantidad total de nombres
```

```python
# Manera 2 de renombrar: usando .rename (no necesita .copy), especificando un diccionario
# con nombre_antiguo : nombre_nuevo
datos09 = datos01.rename(columns={'HV009': 'TOTAL_HOGAR',
                                  'HV010': 'TOTAL_MUJERES',
                                  'HV014': 'TOTAL_NINOS',
                                  'HV015': 'ENTREVISTA',
                                  'HV024': 'DEPARTAMENTO',
                                  'HV025': 'AREA',
                                  'HV035': 'NINOS_ELEG',
                                  'HV042': 'HEMOG_ELEG'})
# no es necesario indicar todos los nombres
```

```python
datos08.equals(datos09)
```

```python
datos08
```

## Creación de variables

```python
# Creamos variable a variable o columna a columna
datos10 = datos08.copy()
datos10['TOTAL_MUJ_NIN'] = datos10['TOTAL_MUJERES'] + datos10['TOTAL_NINOS']
datos10['TOTAL_OTROS']   = datos10['TOTAL_HOGAR'] - datos10['TOTAL_MUJ_NIN']
```

```python
# Creamos las variables con assign (similar al mutate de R)
datos11 = datos08.copy().assign(
    TOTAL_MUJ_NIN = lambda w: w['TOTAL_MUJERES'] + w['TOTAL_NINOS'],
    TOTAL_OTROS   = lambda x: x['TOTAL_HOGAR'] - x['TOTAL_MUJ_NIN'])
```

```python
datos10.equals(datos11)
```

```python
datos11
```

Otro ejemplo:

```python
import numpy as np
datos11 = datos08.copy()
datos11['LIMA'] = np.where(datos11['DEPARTAMENTO'] == "LIMA", "SI", "NO")
```

```python
datos12 = datos08.copy().assign(
    LIMA = lambda x: x['DEPARTAMENTO'].apply(lambda y: "SI" if y == "LIMA" else "NO"))
```

```python
datos11.equals(datos12)
```

## Filtros

```python
# Manera 1 de filtrar: colocando la condición dentro de corchetes
datos19 = datos18[datos18['AREA'] == "URBANO"]
datos19.head()
```

```python
# Manera 2 de filtrar: colocando la condición (entre comillas) en el método query, previa copia
datos20 = datos18.copy().query('AREA == "URBANO"') # query = consulta
datos20.head()
```

```python
datos19.equals(datos20)
```

Otro ejemplo:

Otro ejemplo, con o:

```python
# Manera 1 (o: |) Alt + 124
datos23 = datos18[(datos18['AREA'] == "URBANO") | (datos18['DEPARTAMENTO'] == 4)]
datos23.head()
```

```python
# Manera 2 (o: |, or)
datos24 = datos18.copy().query('AREA == "URBANO" or DEPARTAMENTO == 4')
datos24.head()
```

```python
datos23.equals(datos24)
```
