El archivo U3 - DATOS01.txt contiene 3 columnas del conjunto de datos de la ENDES del año 2022.

```python
import pandas as pd
datos1a = pd.read_csv('U8 - DATOS01.txt', delimiter = '\t', decimal = ',', names = ["CENTRO_POBLADO","LONGITUD","LATITUD"], skiprows = 1)
datos1a
```

```python
datos1a.info()
```

```python
url = 'https://raw.githubusercontent.com/jeguns/Manipulacion-de-datos-con-R-y-Python/main/Python/8%20-%20Lectura%20de%20datos/U8%20-%20DATOS01.txt'
datos1b = pd.read_csv(url, delimiter = '\t', decimal = ',', names = ["CENTRO_POBLADO","LONGITUD","LATITUD"], skiprows = 1)
datos1b.head(10)
```
